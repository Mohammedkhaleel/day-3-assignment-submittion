# Question 2

```python
for prime_num in range(1,1000):
  if (prime_num > 1):
    for i in range(2,prime_num):
      if(prime_num % i) == 0:
        break
    else:
      print(prime_num, 'is a prime number')
```

# Question 4

```python
num = int(input("Enter your desired number:"))
while num > 1:
  for i in range(2,num):
    if (num % i) == 0:
      print(num,"is not a prime number")
      break
    else:
      print(num,"is a prime number")
      break
```
